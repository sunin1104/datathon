### 수면 시간 문자열을 수치형으로 변환  

```python
sleep_map = {
    'Less than 5 hours': 4.5,
    '5-6 hours': 5.5,
    '6-7 hours': 6.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9.0
}
train_df['Sleep Duration'] = train_df['Sleep Duration'].map(sleep_map)
```

### 도시 / 시골 분류 

```python
city_counts = train_df['City'].value_counts()
filtered_city_counts = city_counts[city_counts >= 10].head(30)

print(len(filtered_city_counts))
```

City가 30개까진 충분히 샘플이 많고 >> 그 외 도시들은 너무 적은 샘플 (7명, 5명...) → 통계적으로 신뢰 어려움

```python
top_30_city = train_df['City'].value_counts().head(30).index.tolist()

df_top30 = train_df[train_df['City'].isin(top_30_city)].copy()

df_top30.to_csv("filtered_top30_cities.csv", index=False)
```
