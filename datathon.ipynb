```python
# // 수면 시간 그룹화
def sleep_quality(duration):
    if 6 <= duration <= 8:
        return 'Good'
    else:
        return 'Poor'
    
train_df['SleepQuality'] = train_df['Sleep Duration'].apply(sleep_quality)
```

도시 기준으로 잘 잔 사람 vs 잘 못 잔 사람 우울 점수 

```python
urban_df = train_df[train_df['region'] == 'urban']

urban_good= urban_df[urban_df['SleepQuality']=='Good']['Depression'].dropna()
urban_poor= urban_df[urban_df['SleepQuality']=='Poor']['Depression'].dropna()

from scipy.stats import ttest_ind
t_stat_urban, p_urban = ttest_ind(urban_good, urban_poor, equal_var=False)

print(f"[도시] 잘 잔 사람 vs 못 잔 사람 간 우울 점수 t-test p-value = {p_urban:.4f}")

alpha=0.05

if p_urban < alpha:
    print("유의미한 차이가 있습니다. (귀무가설 기각)")
else:
    print("유의미한 차이가 없습니다. (귀무가설 채택)")
```

시골 기준으로 잘 잔 사람 vs 잘 못 잔 사람 우울 점수 

```python
non_urban_df = train_df[train_df['region'] == 'rural']

non_urban_good=non_urban_df[non_urban_df['SleepQuality']=='Good']['Depression'].dropna()
non_urban_poor=non_urban_df[non_urban_df['SleepQuality']=='Poor']['Depression'].dropna()

t_stat_non_urban, p_non_urban= ttest_ind(non_urban_good,non_urban_poor, equal_var=False)

print(f"[중소도시] 잘 잔 사람 vs 못 잔 사람 간 우울 점수 t-test p-value={p_non_urban:.4f}")

if p_non_urban<alpha:
    print("유의미한 차이가 있습니다. (귀무가설 기각)")
else:
    print("유의미한 차이가 없습니다. (귀무가설 채택)")
```

도시(urban)와 중소도시(non-urban) 간에 Depression(우울 점수)에 차이가 있는가? 

```python
from scipy.stats import ttest_ind

urban_group = train_df[train_df['region'] == 1]['Depression'].dropna()
rural_group = train_df[train_df['region'] == 0]['Depression'].dropna()

t_stat, p_value = ttest_ind(urban_group, rural_group, equal_var=False)

alpha = 0.05
if p_value < alpha:
    print("도시와 중소도시 간 우울 점수에 통계적으로 유의한 차이가 있습니다.")
else:
    print("통계적으로 유의한 차이가 없습니다.")
```

도시/중소도시 나이 분포 비교 

```python
print(train_df['Age'].max())
print(train_df['Age'].min())
```

```python
# 나이 그룹화 함수
def categorize_age(age):
    if age < 20:
        return '10대 이하'
    elif age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    elif age < 50:
        return '40대'
    elif age < 60:
        return '50대'
    else:
        return '60대 이상'

# 새로운 컬럼 추가
train_df['AgeGroup'] = train_df['Age'].apply(categorize_age)
```

```python
import seaborn as sns

order=['10대 이하','20대','30대','40대','50대','60대 이상']

plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='AgeGroup', hue='region', palette='Set2', order=order)

plt.title('도시 vs 중소도시 연령대별 분포', fontsize=15)
plt.xlabel('연령대', fontsize=12)
plt.ylabel('인원 수', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='지역', title_fontsize=12, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
